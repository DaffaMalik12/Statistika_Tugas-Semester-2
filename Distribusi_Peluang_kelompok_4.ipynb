## Distribusi Binomial


```python
# Deklarasi Pustaka
from scipy.stats import binom
import matplotlib.pyplot as plt 
import numpy as np
```

# Contoh Soal 
1. Koin dilemparkan sebanyak 5 kali dengan peluang 1/2 berapakah peluang akan muncul gambar jika koin dilemparkan 3 kali?

```python
# Deklarasisikan nilai n dan p
n= 5
p=0.5
x=3

b = binom(n,p)

print(b.pmf(3)) #peluang mass function, P(x=3)
```

```python
# Membuat Tabel Distribusi Peluang 
x_bin = list(range(n+1))
dist_bin = [b.pmf(x) for x in x_bin]

print("X\tP(x)")
for i in range(n+1):
  print(str(x_bin[i]) + "\t" + str(dist_bin[i]))

```

```python
plt.bar(x_bin , dist_bin)
plt.title("Peluang")
plt.show()
```

## Distribusi HyperGeometrik

```python
# Deklarasi  Pustaka 
from scipy.stats import hypergeom
import matplotlib.pyplot as plt 
import numpy as np
import math
```

1. Sebuah populasi berisi 20 barang, 10 barang diantarannya cacat Diambil sampel sebanyak 4 buah untuk mendekteksi kecacatan barang tersebut, berapakah probabilitas 3 barang dinyatakan cacat? anggaplah sampel diambil tanpa pengembalian 

jawab:

Dik: N=20, s=10, n=4, x=3


```python
# Deklarasikan Nilai Populasi dan Sampel
  
  # jumlah total barang dalam populasi
N = 20

# jumlah barang cacat dalam populasi
s = 10

# jumlah barang yang diambil dalam sampel
n = 4

# jumlah barang cacat yang diambil dalam sampel
x = 3


```

```python
# menghitung probabilitas menggunakan distribusi hipergeometrik
probabilitas = (math.comb(s, x) * math.comb(N-s, n-x)) / math.comb(N, n)

print("Probabilitas 3 barang dinyatakan cacat adalah: {:.4f}".format(probabilitas))
```

```python

# menghitung probabilitas untuk setiap kemungkinan jumlah barang cacat dalam sampel
probabilitas = []
for d in range(n+1):
    p = (math.comb(s, d) * math.comb(N-s, n-d)) / math.comb(N, n)
    probabilitas.append(p)

# menampilkan hasil probabilitas dan grafik distribusi hipergeometrik
fig, ax = plt.subplots()
ax.bar(range(n+1), probabilitas)
ax.set_xlabel('Jumlah Barang Cacat dalam Sampel')
ax.set_ylabel('Probabilitas')
ax.set_title('Distribusi Hipergeometrik')
plt.show()
```

## Distribusi Peluang Kontinu Standar Normal

```python
# Deklarasi Pustaka
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

1. Rata-rata pembalap motoGP menyelesaikan kompetisi selama 45 menit dengan simpangan baku nya 15 , berapa peluang Valentino rossi menyelesaikan balapan dalam waktu kurang dari 30 menit 

```python
# menghitung z-score
mean = 45
std_dev = 15
x = 30
z_score = (x - mean) / std_dev

print(z_score)
```

```python
# menghitung peluang menggunakan distribusi normal standar
prob = stats.norm.cdf(z_score)

# menampilkan hasil
print(f'Peluang Valentino Rossi menyelesaikan balapan dalam waktu kurang dari 30 menit adalah {prob:}')
```

```python
# menampilkan grafik
x_axis = np.arange(-4, 4, 0.001)
y_axis = stats.norm.pdf(x_axis, 0, 1)
plt.plot(x_axis, y_axis)

plt.fill_between(x_axis[x_axis <= z_score], y_axis[x_axis <= z_score], 0, alpha=0.5, color='red')
plt.axvline(x=z_score, linestyle='--', color='red')

plt.title('Distribusi Normal Standar')
plt.xlabel('z-score')
plt.ylabel('PDF')
plt.show()
```

2. Sebuah pabrik memproduksi TV, didapatkan rata-rata 20 TV yang rusak dan standar deviasinya 13. Berapa peluang rusaknya TV jika TV yang diproduksi dibawah 60.

```python
# menghitung z-score
mean = 20
std_dev = 13
x = 60
z_score = (x - mean) / std_dev

print(z_score)
```

```python
# menghitung peluang menggunakan distribusi normal standar
prob = stats.norm.cdf(z_score)

# menampilkan hasil
print(f'Peluang rusaknya TV jika TV yang diproduksi dibawah 60 adalah {prob:}')
```

```python
# menampilkan grafik
x_axis = np.arange(-4, 4, 0.001)
y_axis = stats.norm.pdf(x_axis, 0, 1)
plt.plot(x_axis, y_axis)

plt.fill_between(x_axis[x_axis <= z_score], y_axis[x_axis <= z_score], 0, alpha=0.5, color='red')
plt.axvline(x=z_score, linestyle='--', color='red')

plt.title('Distribusi Normal Standar')
plt.xlabel('z-score')
plt.ylabel('PDF')
plt.show()
```
